# Let us look at the data set a bit differently !!!

## LET US CONSTUCT A FEW NEW FEATURES WHICH WILL HELP US IN OUR ANALYSIS

```python
t=df.iloc[:,5:8]
t.corr()
```

```python
# We can see that these three features are highly co related 
# So why not combine them to make a new feature called Intensity of workout
```

```python
df['Intensity']=0.4*df['Duration']+0.3*df['Heart_Rate']+0.3*df['Body_Temp']
```

## As per the domain knowledge of the team we gave the highest weight to the feature Duration and equal weight to Heart_Rate and Body_Temp

```python
df['BMI']=df['Weight']/((df['Height']/100)**2)
```

## We are adding BMI also USING Weight and Height

```python
df_=df
df_.drop(columns=['Age','Height','Weight','Duration','Heart_Rate','Body_Temp'],inplace=True)
```

```python
df_.head(3)
```

# Calories Analysis

## Intensity Vs Calories

```python
plt.figure(figsize=(4,3))
sns.scatterplot(x=df_['Intensity'],y=df_['Calories'],hue=df['Gender'])
```

#### We can clearly see the trend here , the higher the intensity of the workout the more the calories burnt 

#### One more thing is pretty visible from this graph ,FEMALES WITH SAME INTENSITY OF WORKOUT ARE BURNING LESSER NUMBER OF CALORIES

##### LET US OBSERVE THIS AND ANALYSE IT MORE, IT IS PRETTY CLEAR THAT INTENSITY OF THE WORKOUT IS NOT THE ONLY FACTOR WHICH DECIDES THE AMOUNT OF CALORIES BURNT 

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Create a figure
fig = plt.figure()

# Add a 3D subplot
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df_['Intensity'], df_['BMI'], df_['Calories'], c='r', marker='o')
```

## BMI IS ALSO PLAYING A SIGNIFICANT ROLE IN DETERMINING THE NUMBER OF CALORIES BURNT 

```python
yy=df_.groupby('Gender',as_index=False)['BMI'].mean()
yy.head()
```

```python
plt.figure(figsize=(3,2))
ax=sns.barplot(x=yy['Gender'],y=yy['BMI'])
for i in ax.containers:
    print(ax.bar_label(i))
```

## THIS ALSO EXPLAINS THE FACT THAT FEMALES ARE BURNING SLIGHTLY LESSER CALORIES BECAUSE THEY HAVE COMPARITIVELY LESSER BMI 

## OVERALL CONCLUSION IS INTENSITY AND YOUR BODY ORIENTATION PLAYS A SIGNIFICANT ROLE IN THE AMOUNT OF CALORIES YOU BURN

## FOR THE SAME INTENSITY OF WORKOUT A PERSON WHO IS HEAVIER THAN YOU WILL BE BURNING MORE CALORIES
